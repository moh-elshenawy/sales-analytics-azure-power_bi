```python
from pyspark.sql.functions import col
```

```python
df_customers = df_customers.select(

    col("customer_id").cast("int"),

    col("full_name"),

    col("email"),

    col("country"),

    col( "registration_date").cast("date")

)

df_customers.printSchema()

```

```python
display(df_transactions)
```

```python
df_retails_silver=df_transactions.join(df_customers,'customer_id').join(df_products,'product_id').join(df_store,'store_id')

df_retails_silver.write.mode("overwrite").saveAsTable("databreakkk.sales.retails_silver")
```

```python
from pyspark.sql.functions import sum,count
```

```python
df_sales_by_country=df_retails_silver.groupBy("country").agg(count("transaction_id").alias("total_sales"))
df_sales_by_country.show(5)
df_sales_by_country.write.mode("overwrite").saveAsTable("databreakkk.sales.sales_by_country_gold")
```

```python
df_category_by_country=df_retails_silver.groupBy("country","category").agg(sum("price").alias("total_sales"))
df_category_by_country.show(5)
```

```python
df_category_by_country.write.mode("overwrite").saveAsTable("databreakkk.sales.category_by_country_gold")        
```

```python
df_top_product=df_retails_silver.groupBy("product_name").agg(sum("price").alias("total_sales")).orderBy(col("total_sales").desc())
df_top_product.show(5)
```

```python
df_top_product.write.mode("overwrite").saveAsTable("databreakkk.sales.top_product_gold")
```

```python
df_top_store=df_retails_silver.groupBy("store_name").agg(sum("price").alias("total_sales")).orderBy(col("total_sales").desc())
df_top_store.show(5)
df_top_store.write.mode("overwrite").saveAsTable("databreakkk.sales.top_store_gold")
```

```python
from pyspark.sql.functions import month, year

df_monthly_sales = df_retails_silver \
    .withColumn("year", year(col("transaction_date"))) \
    .withColumn("month", month(col("transaction_date"))) \
    .groupBy("year", "month") \
    .agg(sum("price").alias("total_sales")) \
    .orderBy(col("year"), col("month"))


df_monthly_sales.show(10)
df_monthly_sales.write.mode("overwrite").saveAsTable("databreakkk.sales.monthly_sales_gold")
```

```python
df_top_customer=df_retails_silver.groupBy("full_name").agg(sum("price").alias("total_sales")).orderBy(col("total_sales").desc())
df_top_customer.show(5)
df_top_customer.write.mode("overwrite").option("mergeSchema", "true").saveAsTable("databreakkk.sales.top_customer_gold")
```
